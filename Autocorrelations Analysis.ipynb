```python
DATAdf = pd.read_csv(r"C:\Users\Luke\source\repos\Phi4Project\LatticeDATA.csv", header=None)
DATAdf = DATAdf.drop([len(DATAdf.columns)-1], axis = 1)
DATAdf = DATAdf[:-1]
#DATAdf.to_csv("L16_lamda0.001_k0.330620")
DATAnumpy = DATAdf.to_numpy()
# """ DATAnumpy[config, ;lexi-index] """

D = 3
L = int((len(DATAdf.columns)+1)**(1/D))
DATAnumpyReshape = np.reshape(DATAnumpy, [len(DATAnumpy)]+3*[L])
# """ DATAnumpyReshape[t_markox, x, y, z] """

print(len(DATAdf))
```

### Indexing Order

```python
t_markox = 0
x = 1
y = 2
z = 3

config = DATAnumpyReshape[t_markox]
phi_z = DATAnumpyReshape[t_markox, x, y, z]
```

### Two point-connected Greens function

$$
G(z)  = \langle \phi(0)\phi(z) \rangle
$$


$$
G_C(z)  = G(z) - \langle \phi(0) \rangle \langle \phi(z) \rangle
$$

```python
G_Z = []
for config in range(len(DATAnumpy)):
    G_z = DATAnumpy[config,0]*DATAnumpy[config,:]
    G_Z.append(G_z)
G_Z = np.array(G_Z)
G_Z_samples = G_Z
G_Z = np.mean(G_Z, axis = 0)

plt.plot( [ (G_Z[t]) for t in range(L)], label = 'Greens Function' )
phi_Z = np.mean(DATAnumpy, axis = 0)
Gc_Z = G_Z - phi_Z[0]*phi_Z

plt.plot( [ (Gc_Z[t]) for t in range(L)] , label = 'Connected Greens Function' )
plt.legend(loc = "upper right")
plt.show()
```
